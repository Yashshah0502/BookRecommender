```python
from statistics import correlation

import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

```python
import numpy as np

books['missing value'] = np.where(books['description'].isna(),1,0)
books['age of the books'] = 2025 - books['published_year']
```

```python
column_of_intrest = ['num_pages','missing value', 'age of the books', 'average_rating']

correlation_mat = books[column_of_intrest].corr(method="spearman")
sns.set_theme(style='white')

plt.figure(figsize=(8,6))

heatmap = sns.heatmap(correlation_mat,annot=True,cmap="coolwarm",fmt=".2f",cbar_kws={"label":"spearman Correlation"})

heatmap.set_title("Correlation Heatmap")
plt.show()
```

```python
# books[(books['published_year'].isna() ) | (books['average_rating'].isna()) | (books['description'].isna()) | (books['num_pages'].isna())]
books_missing = books[~(books['published_year'].isna() ) & ~(books['average_rating'].isna()) & ~(books['description'].isna()) & ~(books['num_pages'].isna())]
```

```python
books_missing
```

```python
#Will check the Distnict categories
books_missing['categories'].value_counts().reset_index().sort_values('count', ascending=False)
```

```python
books_missing['Words_in_descrition'] = books_missing['description'].str.split().str.len()
books_missing = books_missing.drop("Missing Descrition",axis=1)
```

```python
books_missing_20_words = books_missing[books_missing['Words_in_descrition'] > 19]
```

```python
books_missing_20_words
```

```python
books_missing_20_words['title_and_subtitle'] = np.where(books_missing_20_words['subtitle'].isna() , books_missing_20_words['title'],
                                  books_missing_20_words[['title','subtitle']].astype(str).agg(': '.join,axis=1 ))
```

```python
books_missing_20_words
```

```python
books_missing_20_words['tagged_description'] = books_missing_20_words[['isbn13','description']].astype(str).agg(' '.join,axis=1 )
```
